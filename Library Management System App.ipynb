```python
from datetime import date  

class ReturnBook(Toplevel):
    def __init__(self):
        Toplevel.__init__(self)
        self.geometry("650x750+550+200")
        self.title("Return Book")
        self.resizable(False, False)
        self.c = c

        
        
        #TOP Frame
        self.topFrame = Frame(self, height = 150, bg = 'white')
        self.topFrame.pack(fill = X)

        #Bottom Frame
        self.bottomFrame = Frame(self, height = 600, bg = '#fcc324')
        self.bottomFrame.pack(fill = X)
        
        #Heading, Image
        self.top_image = PhotoImage(file = r"C:\Users\ASUS\Desktop\Job_Assignments\LMS\return.png")
        top_image_lbl = Label(self.topFrame, image = self.top_image, bg = 'white')
        top_image_lbl.place(x = 120, y = 10)
        heading = Label(self.topFrame, text = ' Return Book ', font = 'arial 22 bold', fg = '#003f8a', bg = 'white')
        heading.place(x = 290, y = 60)
        
         # Member ID
        self.member_id = StringVar()
        self.lbl_member_id = Label(self.bottomFrame, text = "Member ID", font = 'arial 15 bold', fg = 'white', bg = '#fcc324')
        self.lbl_member_id.place(x = 40, y = 40)
        self.ent_member_id = Entry(self.bottomFrame, width = 30, bd = 4)
        self.ent_member_id.insert(0, 'Please enter a member id: ')
        self.ent_member_id.place(x = 150, y = 45)
        
        # Book ID
        self.book_id = StringVar()
        self.lbl_book_id = Label(self.bottomFrame, text = " Book ID ", font = 'arial 15 bold', fg = 'white', bg = '#fcc324')
        self.lbl_book_id.place(x = 40, y = 80)
        self.ent_book_id = Entry(self.bottomFrame, width = 30, bd = 4)
        self.ent_book_id.insert(0, 'Please enter a book id: ')
        self.ent_book_id.place(x = 150, y = 85)
        
        # Date
        self.lbl_date = Label(self.bottomFrame, text = " Date ", font = 'arial 15 bold', fg = 'white', bg = '#fcc324')
        self.lbl_date.place(x = 40, y = 120)
        self.ent_date = Entry(self.bottomFrame, width = 30, bd = 4)
        self.ent_date.insert(0, 'Please enter current date')
        self.ent_date.place(x = 150, y = 125)
        
        # Button
        button = Button(self.bottomFrame, text = 'Return ', command = self.returnBook)
        button.place(x = 280, y = 160)
        
        print(f"Member ID: {member_id}, Book ID: {book_id}")
        sql_query = "SELECT borrow_id FROM borrows WHERE bmember_id = %s AND bbook_id = %s"
        print(f"SQL Query: {sql_query}")
        c.execute(sql_query, (member_id, book_id))

        
    def returnBook(self):
        # Assuming self.ent_member_id and self.ent_book_id are Entry widgets for member_id and book_id
        member_id = self.ent_member_id.get()
        print(member_id)
        book_id = self.ent_book_id.get()
        print(book_id)

        try:
            # Extract numeric IDs from the composite columns
            book_id_parts = book_id.split('-')
            member_id_parts = member_id.split('-')
            
            print(book_id_parts)
            print(member_id_parts)


            if len(book_id_parts) == 2 and len(member_id_parts) == 2:
                bbook_id = book_id_parts[0]
                bmember_id = member_id_parts[0]

                # Check if the book has been returned
                self.c.execute("SELECT * FROM returns WHERE rmember_id = %s AND rbook_id = %s", (bmember_id, bbook_id))
                return_row = self.c.fetchone()

                if return_row:
                    messagebox.showinfo("Info", "Book has already been returned.", icon="info")
                else:
                    # Retrieve borrow record
                    self.c.execute("SELECT borrow_id FROM borrows WHERE bmember_id = %s AND bbook_id = %s", (bmember_id, bbook_id))
                    borrow_row = self.c.fetchone()

                    if borrow_row:
                        borrow_id = borrow_row[0]

                        # Get current date for return_date
                        return_date = date.today()

                    # Insert a record into the returns table
                        self.c.execute("INSERT INTO returns (rbook_id, rmember_id, return_date) VALUES (%s, %s, %s)",
                              (bbook_id, bmember_id, return_date))
                        self.db_connection.commit()

                    # Update book status to available
                        self.c.execute("UPDATE books SET book_status = %s WHERE book_id = %s", (0, bbook_id))
                        self.db_connection.commit()

                        messagebox.showinfo("Success", "Book returned successfully!", icon="info")

                        # Close the ReturnBook window
                        self.destroy()
                    else:
                        messagebox.showerror("Error", "No matching borrow record found.", icon='warning')
            else:
                messagebox.showerror("Error", "Invalid book or member ID format.", icon='warning')

        except Exception as e:
            print(f"Error returning book: {e}")
            messagebox.showerror("Error", f"An error occurred. Please try again.\nDetails: {e}", icon='warning')
```
